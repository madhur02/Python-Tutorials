#### Write a program to read the content of the files present inside testfolder.
**Here test folder contains 2 sub folders Maths and Science. Our program should go inside each folder and try to read the files present insdie it.**

```python
root_folder = r"C:\Users\majain\Desktop\testfolder"
```

```python
# Get the list of Sub folder present inside testfolder
sub_folders = os.listdir(root_folder)
```

```python
for folder in sub_folders: ## this loop is iterating over sub-folders 'Maths and Science '
    sub_folder_path = os.path.join(root_folder,folder) ## Joining testfolder path with Subfolders
    #print(sub_folder_path)
    files = os.listdir(sub_folder_path) ## get the list of files inside sub-folder (Maths or Science)
    print(files)
    for file in files:
        file_path = os.path.join(sub_folder_path,file)
        myfile = open(file_path)
        print(myfile.read()) ## Reading the content of all files
```

## Fetching all files which ends with .txt

```python
import os
items = os.listdir(r"C:\Users\majain\Desktop\csvScript")
folder = r"C:\Users\majain\Desktop\csvScript"
newlist = []
for names in items:
    if names.endswith(".txt"):
        my_file = open(os.path.join(folder,names))
        newlist.append(my_file.read())
```
